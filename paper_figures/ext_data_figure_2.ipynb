```python
pbmc_tiles_33_file = './scarlink_outs/pbmc_10k_down_33/gene_linked_tiles_celltype.csv.gz'
pbmc_tiles_66_file = './scarlink_outs/pbmc_10k_down_66/gene_linked_tiles_celltype.csv.gz'
pbmc_tiles_33_both_file = './scarlink_outs/pbmc_10k_down_33_both/gene_linked_tiles_celltype.csv.gz'
pbmc_tiles_66_both_file = './scarlink_outs/pbmc_10k_down_66_both/gene_linked_tiles_celltype.csv.gz'
pbmc_tiles_33_rna_file = './scarlink_outs/pbmc_10k_down_33_rna/gene_linked_tiles_celltype.csv.gz'
pbmc_tiles_66_rna_file = './scarlink_outs/pbmc_10k_down_66_rna/gene_linked_tiles_celltype.csv.gz'
pbmc_tiles_all_file = './scarlink_outs/pbmc_all_out_10k/gene_linked_tiles_celltype.csv.gz'
hvg_pbmc_file = "./scarlink_outs/pbmc_all_out_10k/hvg.txt"
```

```python
hvg_pbmc = pandas.read_csv(hvg_pbmc_file, sep='\t', header=None)[0].values
```

```python
pbmc_tiles_33 = pandas.read_csv(pbmc_tiles_33_file, sep='\t')
pbmc_tiles_66 = pandas.read_csv(pbmc_tiles_66_file, sep='\t')
pbmc_tiles_33_both = pandas.read_csv(pbmc_tiles_33_both_file, sep='\t')
pbmc_tiles_66_both = pandas.read_csv(pbmc_tiles_66_both_file, sep='\t')
pbmc_tiles_33_rna = pandas.read_csv(pbmc_tiles_33_rna_file, sep='\t')
pbmc_tiles_66_rna = pandas.read_csv(pbmc_tiles_66_rna_file, sep='\t')
pbmc_tiles_all = pandas.read_csv(pbmc_tiles_all_file, sep='\t')
```

```python
def get_corr_df(df):
    df = df[['gene', 'Spearman corr']].drop_duplicates()
    return df
```

```python
pbmc_tiles_33_corr = get_corr_df(pbmc_tiles_33)
pbmc_tiles_66_corr = get_corr_df(pbmc_tiles_66)
pbmc_tiles_33_rna_corr = get_corr_df(pbmc_tiles_33_rna)
pbmc_tiles_66_rna_corr = get_corr_df(pbmc_tiles_66_rna)
pbmc_tiles_33_both_corr = get_corr_df(pbmc_tiles_33_both)
pbmc_tiles_66_both_corr = get_corr_df(pbmc_tiles_66_both)
pbmc_tiles_all_corr = get_corr_df(pbmc_tiles_all)
```

```python
def compare_plots(dfs, hvg):
    names = ['33% sampled ATAC', '66% sampled ATAC', '33% sampled RNA', '66% sampled RNA',
             '33% sampled ATAC+RNA', '66% sampled ATAC+RNA', 'no downsampling']
    fig, ax = plt.subplots(3, 2, sharex=False, sharey=False, figsize=(7, 8.5))
    df = dfs[-1] # pandas.read_csv(files[-1], sep='\t')
    df = df[df['gene'].isin(hvg)]
    print("N genes:", df.shape[0])
    
    idx = 0
    idx1 = 0
    idx2 = 0
    n_genes = [df.shape[0]]
    for df_f in dfs[:-1]: # files[:-1]:
        #df_f = pandas.read_csv(f, sep='\t')
        df_c = df[['gene', 'Spearman corr']].merge(df_f[['gene', 'Spearman corr']], how='left', on='gene')
        df_c = df_c.fillna(0)
        slope, intercept, r_value, pv, se = scipy.stats.linregress(df_c['Spearman corr_x'], df_c['Spearman corr_y'])
        if intercept < 0:
            lbl = "y={0:.3f}x - {1:.3f}".format(slope, abs(intercept))
        else:
            lbl = "y={0:.3f}x + {1:.3f}".format(slope, abs(intercept))
        seaborn.scatterplot(x='Spearman corr_x', y='Spearman corr_y', data=df_c, alpha=0.5, 
                            s=10, marker='.', edgecolor=None, ax=ax[idx1][idx2], label=lbl).legend(loc='best')

        ax[idx1][idx2].set_xlim((-0.05, 0.85))
        ax[idx1][idx2].set_ylim((-0.05, 0.85))
        ax[idx1][idx2].set_xlabel('No downsampling')
        ax[idx1][idx2].set_ylabel(names[idx])
        ax[idx1][idx2].plot((-0.05, 0.85), (-0.05, 0.85), color='black', ls='--')
        ax[idx1][idx2].spines[['right', 'top']].set_visible(False)
        sta, pv = scipy.stats.mannwhitneyu(df_c['Spearman corr_x'].values, df_c['Spearman corr_y'].values)
        print(names[idx], "p-val:", pv)
        idx += 1
        idx1 = int(idx//2)
        idx2 = idx%2
        ax[0][1].set_yticklabels([])
        ax[1][1].set_yticklabels([])
        
    plt.tight_layout()
```

```python
compare_plots(dfs=[pbmc_tiles_33_corr, pbmc_tiles_66_corr, 
                   pbmc_tiles_33_rna_corr, pbmc_tiles_66_rna_corr,
                   pbmc_tiles_33_both_corr, pbmc_tiles_66_both_corr,
                   pbmc_tiles_all_corr], hvg=hvg_pbmc)
plt.savefig('./tmp/downsample_corr_plots.pdf', transparent=True)
```
