```r
library(ggplot2)
library(gtable)
library(gridExtra)
library(grid)
library(patchwork)
```

```r
pbmc_tbl <- read.table("./scarlink_outs/pbmc_all_out_10k/cell_info.txt", sep='\t', header=TRUE)
panc_tbl <- read.table("./scarlink_outs/panc_p1_all_out_10k/cell_info.txt", sep='\t', header=TRUE)
pit_tbl <- read.table("./scarlink_outs/pituitary_all_out_10k/cell_info.txt", sep='\t', header=TRUE)
bmmc_tbl <- read.table("./scarlink_outs/bmmc_all_out_10k/cell_info.txt", sep='\t', header=TRUE)
cortex_tbl <- read.table("./scarlink_outs/human_cortex_all_out_10k/cell_info.txt", sep='\t', header=TRUE)
mouse_skin_tbl <- read.csv("./scarlink_outs/mouse_skin_all_out_10k/cell_info.txt", sep='\t', header=TRUE)

pbmc_tbl$orig.ident <- 'PBMC'
panc_tbl$orig.ident <- 'Pancreas'
pit_tbl$orig.ident <- 'Pituitary'
bmmc_tbl$orig.ident <- 'BMMC'
cortex_tbl$orig.ident <- 'Cortex'
mouse_skin_tbl$orig.ident <- 'Mouse skin'
```

```r
options(repr.plot.width = 15, repr.plot.height = 11)

p1 <- ggplot(pbmc_tbl, aes(x=orig.ident, y=ReadsInTSS)) + geom_violin() + ylim(0, 20000) + geom_boxplot(width=.1)
p2 <- ggplot(panc_tbl, aes(x=orig.ident, y=ReadsInTSS)) + geom_violin() + ylim(0, 20000) + geom_boxplot(width=.1)
p3 <- ggplot(pit_tbl, aes(x=orig.ident, y=ReadsInTSS)) + geom_violin() + ylim(0, 20000) + geom_boxplot(width=.1)
p4 <- ggplot(bmmc_tbl, aes(x=orig.ident, y=ReadsInTSS)) + geom_violin() + ylim(0, 20000) + geom_boxplot(width=.1)
p5 <- ggplot(cortex_tbl, aes(x=orig.ident, y=ReadsInTSS)) + geom_violin() + ylim(0, 20000) + geom_boxplot(width=.1)
p6 <- ggplot(mouse_skin_tbl, aes(x=orig.ident, y=ReadsInTSS)) + geom_violin() + ylim(0, 20000) + geom_boxplot(width=.1)
p1+p5+p4+p2+p3+p6
```

```r
mean(pbmc_tbl$ReadsInTSS)
mean(panc_tbl$ReadsInTSS)
mean(pit_tbl$ReadsInTSS)
mean(bmmc_tbl$ReadsInTSS)
mean(cortex_tbl$ReadsInTSS)
mean(mouse_skin_tbl$ReadsInTSS)
```
